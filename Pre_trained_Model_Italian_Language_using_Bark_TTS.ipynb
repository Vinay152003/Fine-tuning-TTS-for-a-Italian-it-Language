```python
# Install Bark (if not already installed)
!pip install git+https://github.com/suno-ai/bark.git

from bark import generate_audio, SAMPLE_RATE
from scipy.io.wavfile import write as write_wav
import numpy as np

# Define the text and language/speaker preset
text_prompt = "Il servizio di oogle, offerto gratuitamente, traduce istantaneamente parole, frasi e pagine web dall'inglese e oltre 100 altre lingue."
voice_preset = "v2/it_speaker_2"  # Adjust this as needed

# Generate the audio using Bark's `generate_audio` method
audio_array = generate_audio(text_prompt, history_prompt=voice_preset)

# Convert the audio array to a NumPy array and save it as a .wav file
audio_array = np.array(audio_array)
write_wav("bark_output.wav", SAMPLE_RATE, audio_array)

# Play the generated audio
from IPython.display import Audio
Audio(audio_array, rate=SAMPLE_RATE)
```
